```python
import urllib2
from bs4 import BeautifulSoup
```

```python
articleURL="https://kafka.apache.org/documentation"
```

```python
page=urllib2.urlopen(articleURL).read().decode('utf8','ignore')
soup=BeautifulSoup(page,'lxml')
soup
```

```python
soup.find('p').text
```

```python
text=' '.join(map(lambda p:p.text,soup.find_all('p')))
```

```python
text
```

```python
text.encode('ascii',errors='replace').replace("?"," ")
```

```python
text=''.join([i if ord(i) < 128 else ' ' for i in text])
```

```python
text
```

```python
import nltk
from nltk.tokenize import word_tokenize,sent_tokenize
from nltk.corpus import stopwords
from string import punctuation
```

```python
custStopWords=set(stopwords.words('english')+list(punctuation))
custStopWords
```

```python
wordlist=[]
text=text.lower()
for word in word_tokenize(text):
    if word not in custStopWords:
        wordlist.append(word);
    
        
```

```python
wordlist
```

```python
 sents=sent_tokenize(text)
```

```python
sents
```

```python
from nltk.probability import FreqDist
freq=FreqDist(wordlist)
freq
```

```python
from heapq import nlargest
```

```python
nlargest(10,freq,key=freq.get)
```

```python
from collections import defaultdict
ranking=defaultdict(int)

for i,sent in enumerate(sents):
    for w in word_tokenize(sent):
        if w in freq:
            ranking[i]+=freq[w]
            
ranking
```

```python
sents_idx=nlargest(3,ranking,key=ranking.get)
sents_idx
```

```python
[sents[j] for j in sorted(sents_idx)]
```
